```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
# import sklearn
```

```python
data=pd.read_csv('diamonds.csv')
data
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
data=data.drop('Unnamed: 0',axis=1)
data
```

```python
data.dtypes
```

```python
data.cut.unique()
```

```python
data.clarity.unique()
```

```python
cut={'Ideal':5, 'Premium':4, 'Very Good':3, 'Good':2, 'Fair':1 }
color={'D':7, 'E':6, 'F':5, 'G':4, 'H':3, 'I':2, 'J':1}
clarity= {'I1':1, 'SI2':2, 'SI1':3, 'VS2':4, 'VS1':5, 'VVS2':6, 'VVS1':7, 'IF':8}
```

```python
data['cut']= data['cut'].map(cut)
```

```python
data['color']=data['color'].map(color)
data['clarity']= data['clarity'].map(clarity)
```

```python
data.dtypes
```

```python
x=data.drop('price',axis=1)
y=data['price']
```

```python
x
```

```python
y
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
#from sklearn.tree import DecisionTreeRegressor
# base_estimator = none : DecisionTreeRegressor with max_depth=1
# model= AdaBoostRegressor(n_estimators=150, base estimator=None, random_state=12)

adaboost = AdaBoostRegressor(n_estimators= 150, base_estimator = RandomForestRegressor(n_estimators=5),random_state=12)
```

```python
adaboost.fit(x_train,y_train)
```

```python
adaboost.score(x_test, y_test)
```
