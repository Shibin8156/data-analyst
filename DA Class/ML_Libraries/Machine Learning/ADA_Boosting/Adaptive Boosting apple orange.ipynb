```python
import pandas as pd
from sklearn.model_selection import train_test_split
from  sklearn.metrics import confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
```

```python
data= pd.read_csv('apples_and_oranges.csv')
```

```python
data.head()
```

```python
x=data.drop('Class',axis=1)
y=data['Class']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.25, random_state=341)
```

### learning_rate = weight applied to each classifier at each boosting iteration . A higher learning rate increases the contribution of  each classifier.

```python
adaboost = AdaBoostClassifier(n_estimators= 50, learning_rate=0.3, random_state=1)

# base_estimator= None : Desicion Tree Classifier with Max_depth=1
# learning_rate is the contribution of each model. small Lr means learning happenms slowly
```

```python
# from sklearn.ensemble import RandomForestClassifier 
#adaboost= AdaboostClassifier(n_estimator= 50, base_estimator= RandomForestClassifier(n_estrimator=50,learning_rate=0.3, random_state=1))
```

```python
adaboost.fit(x_train,y_train)
```

```python
adaboost.score(x_test, y_test)
```
