```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from word2number import w2n
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
data=pd.read_csv('Automobile_data.csv')
```

```python
data
```

```python
data.dtypes
```

```python
data.drop('index',axis=1)
```

```python
data.isna().sum()
```

```python
price_mean=data['price'].mean()
price_mean
```

```python
data['price']=data['price'].fillna(price_mean)
```

```python
data
```

```python
data['num-of-cylinders']=data['num-of-cylinders'].apply(w2n.word_to_num)
```

```python
le_company=LabelEncoder()
le_bodystyle=LabelEncoder()
le_enginetype=LabelEncoder()

data['company']=le_company.fit_transform(data['company'])
data['body-style']=le_bodystyle.fit_transform(data['body-style'])
data['engine-type']=le_enginetype.fit_transform(data['engine-type'])
```

```python
data=data.drop('index', axis=1)
```

```python
data
```

```python
x=data.drop('price',axis=1)
y=data['price']
```

```python
x
```

```python
y
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=2)
```

```python
Gdr= GradientBoostingRegressor(n_estimators=750, learning_rate=0.05, random_state=100)
```

```python
Gdr.fit(x_train,y_train)
```

```python
Gdr.score(x_test,y_test)
```
