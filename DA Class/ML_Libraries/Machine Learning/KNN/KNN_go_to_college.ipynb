```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
data=pd.read_csv('go_to_college.csv')
data
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
a= LabelEncoder()
data['type_school'] =a.fit_transform(data['type_school'])
b= LabelEncoder()
data['school_accreditation'] =b.fit_transform(data['school_accreditation'])
c= LabelEncoder()
data['gender'] =c.fit_transform(data['gender'])
d= LabelEncoder()
data['interest'] =d.fit_transform(data['interest'])
e= LabelEncoder()
data['residence'] =e.fit_transform(data['residence'])
f= LabelEncoder()
data['parent_was_in_college'] =f.fit_transform(data['parent_was_in_college'])
```

```python
data
```

```python
x=data.drop('in_college', axis=1)
y=data['in_college']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.25, random_state=20)
```

```python
model=KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
model2=KNeighborsClassifier(n_neighbors=25)
```

```python
model2.fit(x_train_scaled,y_train)
```

```python
model2.score(x_test_scaled,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
sbn=confusion_matrix(y_test,y_pred)
sbn
```

```python
print(classification_report(y_test,y_pred))
```
