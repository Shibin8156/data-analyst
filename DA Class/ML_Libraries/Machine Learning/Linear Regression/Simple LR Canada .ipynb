```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data1=pd.read_csv('canada_per_capita_income.csv')
data1
```

```python
plt.xlabel('Year')
plt.ylabel('Per capita income')
plt.scatter(data1.year, data1['per capita income (US$)'], color="blue", marker='*')
```

```python
x=data1.drop('per capita income (US$)', axis=1)  # Feature
y=data1['per capita income (US$)']  #Target
```

```python
x
```

```python
y
```

```python
model1=LinearRegression()
model1.fit(x,y)
```

```python
model1.coef_
```

```python
model1.intercept_
```

```python
y_pred=model1.predict(x)
```

```python
pd.DataFrame(({'True y values': y, 'Predicted y values': y_pred}))
```

```python
plt.scatter(x,y)
plt.xlabel('Year')
plt.ylabel('Per capita income')
plt.plot(data1['year'], y_pred,color= 'red')
```
