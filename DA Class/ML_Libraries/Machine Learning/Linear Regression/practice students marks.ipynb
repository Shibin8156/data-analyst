```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('Student_Marks.csv')
data
```

```python
data.isna().sum()
```

```python
a=data.number_courses.median()
b=data.time_study.median()
c=data.Marks.median()
```

```python
data['number_courses'] =data['number_courses'].fillna(a)
data['time_study'] =data['time_study'].fillna(b)
data['Marks'] =data['Marks'].fillna(c)
```
