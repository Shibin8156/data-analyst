```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("healthcare-dataset-stroke-data.csv")
data
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.describe()
```

```python
data['bmi'].fillna(data['bmi'].mean(), inplace=True)
data.isnull().sum()
```

```python
data
```

```python
data['stroke'].value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data["age"], bins=10, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="gender",data=data)
plt.title("Gender Count")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(x="smoking_status",y="bmi",data=data)
plt.title("BMI Distribution by Smoking Status")
plt.xlabel("Smoking Status")
plt.ylabel("BMI")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="stroke",data=data)
mean_stroke =data["stroke"].mean()
plt.axhline(mean_stroke, color='red',linestyle='--',label='Mean')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
data["smoking_status"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Smoking Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x="age", y="avg_glucose_level",data=data)
plt.title("Age vs. Average Glucose Level")
plt.xlabel("Age")
plt.ylabel("Average Glucose Level")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(),annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
y_pred=model.predict(x)
```

```python
plt.axline([0,0],[1,1],c= 'brown')
plt.scatter(y,y_pred)
```

```python
model.score(x,y)
```
