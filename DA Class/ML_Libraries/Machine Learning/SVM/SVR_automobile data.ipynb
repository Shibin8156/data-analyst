```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from word2number import w2n
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('Automobile_data.csv')
data
```

```python
data.head()
```

```python
data=data.drop('index', axis=1)
data
```

```python
a= LabelEncoder()
data['company'] =a.fit_transform(data['company'])
b= LabelEncoder()
data['body-style'] =a.fit_transform(data['body-style'])
c= LabelEncoder()
data['engine-type'] =a.fit_transform(data['engine-type'])
```

```python
data
```

```python
data.isna().sum()
```

```python
data['price'].fillna(data['price'].mean(), inplace=True)
data.isnull().sum()
```

```python
data['num-of-cylinders']=data['num-of-cylinders'].apply(w2n.word_to_num)
```

```python
data
```

```python
x=data.drop('price',axis=1)
y=data['price']
```

```python
data
```

```python
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.25, random_state=12)
```

```python
x
```

```python
y
```

```python
svm=SVR(kernel='linear')
svm.fit(x_train, y_train)
svm.score(x_test, y_test)
```

```python
svm_poly=SVR(kernel='poly', degree=4, C=2.5, gamma='auto') #scale
svm_poly.fit(x_train, y_train)
svm_poly.score(x_test, y_test)
```

```python
svm_gauss =SVR(kernel='rbf') #default kernal --- gaussion
svm_gauss.fit(x_train, y_train)
svm_gauss.score(x_test, y_test)
```

```python
svm_sig =SVR(kernel='sigmoid')
svm_sig.fit(x_train, y_train)
svm_sig.score(x_test, y_test)
```
