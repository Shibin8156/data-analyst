## Group by

  ### Creates sub dataframe based on single coloumn value or a combination of multiple column values

```python
df1=pd.DataFrame({'Name': ['A','B','C','D','E'],
                 'Sex': ['M','F','F','M','F',],
                 'Age':[23,21,34,28,32]})
df1
```

```python
grp= df1.groupby('Sex')
grp
```

```python
grp.apply(display)
```

```python
data=pd.read_csv("weather_data.csv")
data
```

```python
grp=data.groupby('city')
grp
```

```python
grp.apply(display)
```

```python
# using loop
for city,city_df in grp:
    print(city)
    print(city_df)
```

```python
grp.get_group("paris")
```

```python
# getting minimum values in each column
grp.min()
```

```python
grp.max()
```

```python
grp.get_group('paris').max()
```

### Concat(

```python
tcs={'id':[101,102,103,104],
    'name':['Max','Dipin','Jose','Rahul'],
    'Age':[25,26,27,29]}

wipro={'id':[101,102,103,104],
    'Name':['Dev','Sreenath','Sreerag','Shafeel'],
    'Age':[24,24,27,26]}
```

```python
tcs_emp=pd.DataFrame(tcs)
wipro_emp=pd.DataFrame(wipro)
```

```python
tcs_emp
```

```python
wipro_emp
```

```python
pd.concat([tcs_emp,wipro_emp])
```

```python
pd.concat([tcs_emp,wipro_emp], ignore_index=True)
```

```python
pd.concat([tcs_emp,wipro_emp], keys=['TCS','WIPRO'])
```

```python
#  contact horizontally
pd.concat([tcs_emp,wipro_emp], axis=1)
```

```python
# Concatenating a Dataframe with a series
salary=pd.Series([65000,75000,15000,25000], name= 'Salary')
pd.concat([tcs_emp,salary], axis=1)
```

## merge()

```python
tcs={'id':[101,102,103,106],
    'name':['Max','Dipin','Jose','Rahul'],
    'age':[25,26,27,29]}

wipro={'id':[101,102,103,105],
    'name':['Dev','Sreenath','Sreerag','Shafeel'],
    'age':[25000,34000,37000,20000]}
tcs_emp=pd.DataFrame(tcs)
wipro_emp=pd.DataFrame(wipro)
```

```python
tcs_emp
```

```python
# inner merging --- series keys common to both the dataframe
pd.merge(tcs_emp,wipro_emp, on='id')
```

```python
pd.merge(tcs_emp,wipro_emp, on='id',suffixes=['_left','_right'])
```

```python
pd.merge(tcs_emp,wipro_emp, on='id', how='outer')    #how= 'inner' by default
```

```python
pd.merge(tcs_emp,wipro_emp, on='id', how='outer', indicator=True)
```

```python
pd.merge(tcs_emp,wipro_emp, on='id', how='left')
```

```python
pd.merge(tcs_emp,wipro_emp, on='id', how='right')
```
