### skip unnecessary headers and add new headers while reading csv file

```python
data=pd.read_csv('stock_data.csv')
data
```

```python
# skiprows skips specific no of rows from begining
data=pd.read_csv('weather_data.csv',skiprows=2)
data
```

```python
# set the specified row as the header discard the remaining rows of the
# dataframe

data2= pd.read_csv('stock_data.csv',header=3)
data2
```

```python
# makes the reader a new header
d=pd.read_csv('stock_data.csv',header=None)
d
```

```python
# set new name for the columns

data3=pd.read_csv('stock_data.csv',header= None, names=['a','b','c','d','e'])
data3
```

```python
# selects only the first nrows from the dataframe

data4=pd.read_csv('stock_data.csv', nrows=3)
data4
```

```python
#  SET specific specific values from the dataframe as null values
# NaN - Not a Number (missing values)
 
data5= pd.read_csv('stock_data.csv', na_values = ['n.a.','not available',-1])
data5
```

```python
data6= pd.read_csv('stock_data.csv', na_values= {'eps':['not available',-1],
                                                'revenue':['n.a.',-1],
                                                'price':['n.a.',-1],
                                                'people':['n.a.','not vailable',-1]})
data6
```

```python
# converter method 

def convert_people(name):
    if name=='larry page':
        return 'Ben'
    else:
        return name
    
data_convert = pd.read_excel('stock_data.xlsx',
                            converters={'people':convert_people})
data_convert
```
