## Handling missing values

```python
data_missing = pd.read_csv('weather_data_missing.csv')
data_missing
```

```python
# checking for missing values
data_missing.isna()
```

```python
data_missing.isnull()
```

```python
data_missing.isna().sum()
```

```python
data_missing.isnull().sum()
```

```python
# fill missing values
data2 = data_missing.fillna(0)
data2
```

```python
avg= data_missing['temperature'].mean()
data_missing.fillna(avg)
# to save the original dataframe should use- "data_missing=data_missing.fillna(avg)""
```

```python
# Forward fill
data_missing.fillna(method='ffill') 
```

```python
# backward fill
data_missing.fillna(method='bfill')
```

```python
#  Linear interpolation : fills with average of preceeding and succeeding values
# 30.2+28.0/2= 30.0

data_missing.interpolate()
```

```python
# drop rows with null values
data_missing.dropna()
```

```python
# drop rows with full NaN values

data_missing.dropna(how='all')
```

```python
# 
data_missing.dropna(how='any') 
```

## Replace

```python
data= pd.read_csv('missing_weather_data.csv')
data
```

```python
data.replace('-99999', np.NaN)
```

```python
data.replace(['-99999','-88888'], np.NaN)
```

```python
data.replace({'temperature':'-99999','windspeed':'-88888','event':'0'}, np.NaN)
```

```python
data.replace({'temperature':'-99999','windspeed':['-88888','-9999'],'event':'0'}, np.NaN)
```

```python
data.replace({'-9999':np.NaN, '-88888':np.NaN, '0':'Rain'})
```

```python
# regular expression
# to replce F,C and mph,kmph in this data
# using this '' replace with space

data.replace({'temperature': '[A-Za-z]', 'windspeed': '[A-Za-z]','event':'[A-Za-z]'},'', regex=True)
```

```python
data.replace(['Rain','Sunny','Cloudy','Snow'], [100,101,102,103])
```
