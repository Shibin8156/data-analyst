```python
# Subplot

a=[10,20,30,40,50]
b=[1,2,3,4,5]
c=[2,3,6,8,10]
d=[3,2,1,0,6]
plt.subplot(2,2,1) # also we can use plt.subplot(221)
plt.plot(a,b)
plt.subplot(2,2,2) # also we can use plt.subplot(222)
plt.plot(b,c)
plt.subplot(2,2,3) #  also we can use plt.subplot(223)
plt.plot(a,c)
plt.subplot(2,2,4) #  also we can use plt.subplot(224)
plt.plot(d,b)

plt.xlabel('x values')
plt.ylabel('y values')
```

### Bar Graph 

```python
x_val=[',Shyam','Aswin','Raju','Appu','Jayan','Shibin']
y_val=[86,75,76,74,72,85]
plt.bar(x_val, y_val , color='green')
plt.title('Person Weight Diagram')
plt.xlabel('Name')
plt.ylabel('Weight')
```

```python
# Multiple Plotting(must set bar width here)
import numpy as np
x_val=[',Shyam','Aswin','Raju','Appu','Jayan','Shibin']
x_range=np.arange(len(x_val))   #(0,1,2,3,4,5)
width=0.4
y_val=[86,75,76,74,72,85]



plt.bar(x_range-width,y_val,width=width, label='Current Weight', color='brown')
y_val2=[80,70,72,70,70,80]
plt.bar(x_range,  y_val2, width=width, label='Desired Weight',  color='pink')

plt.xticks(ticks=x_range,labels=x_val)

plt.grid(True)
plt.title('Person Weight Diagram')
plt.xlabel('Name')
plt.ylabel('Weight')
plt.legend()
```

```python
# Horizontal Bar Graph 
x_val=[',Shyam','Aswin','Raju','Appu','Jayan','Shibin']
y_val=[86,75,76,74,72,85]
plt.barh(x_val, y_val , color='green')

plt.title('Person Weight Diagram')
plt.xlabel('Name')
plt.ylabel('Weight')
```

### Pie Chart

```python
slices=[30,40,20]
labels=['men','women','kids']
plt.style.use('classic')
plt.pie(slices,labels=labels)
plt.title('PIE CHART')
```

```python
# Setting color and wedge properties 
colors=['red','blue','pink']

slices=[30,40,20]
labels=['men','women','kids']
plt.style.use('classic')
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'linewidth':2,'edgecolor':'yellow'})
plt.title('PIE CHART')
```

```python
# Explodes and Shadows

slices=[30,40,20]
labels=['men','women','kids']
colors=['red','blue','pink']

# Explodes values
explodes=[0,0,0.1]

plt.pie(slices,labels=labels,colors=colors,wedgeprops={'linewidth':2,'edgecolor':'black'},explode=explodes,shadow=True)
plt.title('PIE CHART')
```

```python
# start angle and percentage value
slices=[30,40,20]
labels=['men','women','kids']
colors=['red','blue','pink']


explodes=[0,0,0.1]

plt.pie(slices,labels=labels,colors=colors,wedgeprops={'linewidth':2,'edgecolor':'black'},
        explode=explodes, shadow=True,
       startangle=90,
       autopct='%.2f%%')
plt.title('PIE CHART')
```

```python
# size, color, marker,edgecolor
import numpy as np
a=np.random.randint(1,10, size=9)
b=np.random.randint(2,15, size=9)


plt.style.use('seaborn-v0_8-muted')
plt.scatter(a,b, s=150, color='pink', marker='o', edgecolors='black', linewidth=2)
```

```python
# cmap and colorbar
a=np.random.randint(1,10, size=9)
b=np.random.randint(2,15, size=9)
plt.style.use('seaborn-v0_8-muted')
colors=b

plt.scatter(a,b, s=100, c=colors, marker='s', edgecolors='black', linewidth=2,
          cmap='Purples')
cbar=plt.colorbar()
cbar.set_label('color variation')
```

```python
# Marker Size

a=np.random.randint(1,10, size=9)
b=np.random.randint(2,15, size=9)
plt.style.use('seaborn-v0_8-muted')
colors=b

# b *10
sizes=b*20

plt.scatter(a,b, s=sizes, c=colors, marker='s', edgecolors='black', linewidth=2,
          cmap='Purples')
cbar=plt.colorbar()
cbar.set_label('Color variation')
```
