```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('Iris.csv')
data.head()
```

### Line Plot
    

```python
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
```

### Styling Plots
#### set_style()    5 themes available (darkgrid,whitegrid,dark,white,ticks)

```python
sns.set_style('white')
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
```

```python
sns.set_style('dark')
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
```

```python
sns.set_style('ticks')
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
```

```python
sns.set_style('darkgrid')
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
```

```python
# Despine(Remove Borders)
sns.set_style('ticks')
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
sns.despine()
```

```python


sns.set_style('ticks')
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
sns.despine(left=True)
```

```python
# Size
plt.figure(figsize=(3,3))
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
```

### Color Palette

```python
# palplot creates a plot for the colors of the palette
palette=sns.color_palette()
sns.palplot(palette)
```

```python
sns.palplot(palette,4)
```

```python
# color shades->  deep, muted, bright, pastel, dark, colorblind
palette=sns.color_palette('dark')
sns.palplot(palette)
```

```python
colors=['orange','white','green']
palette=sns.color_palette(colors)
sns.palplot(palette)
```

```python
# applying color paltte to our plot
sns.set_palette('winter_r')
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
plt.title('Sepal Lenght v/s Sepal Width')
```
