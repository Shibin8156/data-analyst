```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Iris.csv')
```

```python
# scatterplot
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',data=data)
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',data=data,hue='SepalWidthCm')
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',data=data,markers='*',hue='SepalWidthCm')
```

```python
# palette
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',data=data,markers='*',hue='SepalWidthCm',
               palette='gist_earth_r')
```

```python
# legend false
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',data=data,markers='*',hue='SepalWidthCm', legend=False)
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',data=data,markers='*',hue='SepalWidthCm', legend='full')
```

```python
# opacity(alpha-0->1)
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',data=data,markers='*',hue='SepalWidthCm',
               palette='gist_earth_r',alpha=0.5)
```

```python
titanic=pd.read_csv('titanic.csv')
titanic.head()
```

```python
sns.barplot(x='Pclass',y='Fare',data=titanic)
```

```python
# use hue to plot 3 items
sns.barplot(x='Pclass',y='Fare',hue='Sex',data=titanic)
```

```python
# horizontal barplot
sns.barplot(x='Fare',y='Sex',data=titanic)
```

```python
# plot bars in a given order
sns.barplot(x='Pclass',y='Fare',order=[3,1,2],data=titanic)
```

```python
# all bars in a single color
sns.barplot(x='Pclass',y='Fare',data=titanic,color='salmon')
```

```python
# countplot
sns.countplot(x='Sex',data=titanic)
```

```python
sns.countplot(x='Sex',hue="Pclass",data=titanic)
```

```python
# to change plot to y axis set x as y
sns.countplot(y='Sex',hue="Pclass",data=titanic,palette='Purples')
```

```python
# Pairplot
sns.pairplot(titanic,hue='Sex')
```

```python
df=np.array([[3,1,6,9],[7,4,8,3]])
df=pd.DataFrame(df,columns=['A','B','C','D'])
df
```

```python
# Heatmap
sns.heatmap(df,annot=True)
plt.xlabel('Columns')
plt.ylabel('Rows')
```
