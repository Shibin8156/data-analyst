### Reading a csv file

```python
# skiprows skips specific no of rows from begining
data=pd.read_csv('weather_data.csv',skiprows=2)
data
```

```python
# set the specified row as header discarding the remaining rows
data2=pd.read_csv('weather_data.csv',header=3)
data2
```

```python
# makes the header a new record
d=pd.read_csv('weather_data.csv',header=None)
d
```

```python
# sets the name for columns
d=pd.read_csv('weather_data.csv',header=None,names=['a','b','c','d','e'])
d
```

```python
# select only first nrows from dataframe
d=pd.read_csv('weather_data.csv',nrows=3)
d
```

```python
# sets specific values from dataframe as null values
# NaN-Not a Number 
d=pd.read_csv('stock_data.csv',na_values=['n.a.','not available',-1])
d
```

```python
# to set specific cols values as null

d=pd.read_csv('stock_data.csv',na_values={'eps':['not available',-1],
                                          'revenue':['n.a.',-1],
                                          'price':['n.a.',-1],
                                          'people':['n.a.','not available',-1]})
d
```

```python
# Converter method
def convert_people(name):
    if name=='larry page':
        return 'Ben'
    else:
        return name
    
data_convert=pd.read_excel('stock_data.xlsx',converters={'people':convert_people})
data_convert
```

```python
data_convert.to_excel('converted_data.xlsx')
```
