### Group By
###### create sub dataframes on a sible column val or combination of multi col values

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Name':['A','B','C','D','E'],
                "Sex":['M','F','F','M','F'],
                "Age":[23,21,34,28,32]})
df
```

```python
grp=df.groupby('Sex')
grp
```

```python
grp.apply(display)
```

```python
data=pd.read_csv('weather_data.csv')
data
```

```python
grouped_result=data.groupby('city')
print(grouped_result)
```

```python
for city,city_df in grouped_result:
    print(city)
    print(city_df)
```

```python
# get specific groups
grouped_result.get_group('paris')
```

```python
grouped_result.min()
```

```python
grouped_result.get_group('paris').max()
```

```python
# groupby using multiple cols
grp=data.groupby(['city','event'])
grp.apply(display)
```

```python
grp.get_group(('newyork','Sunny'))
```

## Concatt

```python
tcs={'id':[101,102,103,104],
    'name':['Ramu','Raju','Jayan','Arjun'],
    'Age':[15,21,15,36]}

wipro={'id':[101,102,103,104],
    'Name':['Jayan','Arjun','Ramu','Hari',],
    'Age':[26,26,28,27]}
```

```python
tcs_emp=pd.DataFrame(tcs)
wipro_emp=pd.DataFrame(wipro)
```

```python
tcs_emp
```

```python
wipro_emp
```

```python
pd.concat([tcs_emp,wipro_emp])
```

```python
pd.concat([tcs_emp,wipro_emp],ignore_index=True)
# get common index after concatenation
```

```python
pd.concat([tcs_emp,wipro_emp],keys=['TCS','WIPRO'])
```

```python
pd.concat([tcs_emp,wipro_emp],axis=1)
```

```python
# concatenating a dataframe with a series 
salary=pd.Series([65000,75000,45000,36000],name='salary')
pd.concat([tcs_emp,salary],axis=1)
```

### Merge

```python
tcs={'id':[101,102,103,106],
    'name':['Ramu','Raju','Jayan','Arjun'],
    'Age':[15,21,15,36]}

wipro={'id':[101,102,103,105],
    'name':['Jayan','Arjun','Ramu','Hari',],
    'Salary':[65000,75000,45000,36000]}
```

```python
tcs_emp=pd.DataFrame(tcs)
wipro_emp=pd.DataFrame(wipro)
```

```python
tcs_emp
```

```python
wipro_emp
```

```python
pd.merge(tcs_emp,wipro_emp,on='id')
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',suffixes=['_left','_prefix'])
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='outer')   #inner by default 
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='outer',indicator=True)
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='left')
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='right')
```

### Cross tab or Contigency table
###### Computes a freq table

```python
data=pd.read_csv('loan_status.csv')
data
```

```python
pd.crosstab(data.grade,data.loan_status)
```

```python
pd.crosstab(data.grade,data.loan_status,margins=True)  #Sum
```

```python
cross_tab=pd.crosstab(data.grade,[data.loan_status,data.sub_grade])
cross_tab.columns
```

```python
cross_tab
```

```python
cross_tab=pd.crosstab(data.grade,[data.loan_status,data.sub_grade],margins=True)
cross_tab
```

### Map

```python
# Is a python bilt in func used to apply func on each element of a sequence.it can be used on pandas dataframe also
```

```python
# Example 
def square(x):
    return x**2

my_list=[6,5,4,2]
list(map(square,my_list))
```

```python
data=pd.read_csv('weather_data.csv')
data
```

```python
# Square the val of temperature
def square(x):
    return x**2
data['temperature'].map(square)
```

```python
data_copy=data.copy()
data_copy['temp_square']=data['temperature'].map(square)
```

```python
data_copy
```

```python
# for handling multi cols simultaneously 
data[['temperature','windspeed']].applymap(square)
```

```python
data[['temperature','windspeed']].applymap(lambda a:a**2)
```
