```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('iris.csv')
data.head()
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data, hue='SepalLengthCm')
```

```python
# alpha defines the opacity of the points[0,1]
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data, palette="gist_earth_r",hue='SepalLengthCm', alpha=0.5)
```

```python
data1=pd.read_csv('titanic.csv')
data1
```

```python
sns.barplot(x='Pclass', y='Fare', data=data1)
```

```python
sns.barplot(x='Pclass', y='Fare',hue='Sex', data=data1)
```

```python
# horizontal barplot
sns.barplot(x='Fare', y='Sex', data=data1)
```

```python
# plot bars in given order
sns.barplot(x='Pclass', y='Fare',order=[3,1,2], data=data1)
```

```python
#  all bars in single color
sns.barplot(x='Pclass', y='Fare', data=data1, color="salmon")
```

## heatmap

```python
df= np.array([[3,1,6,9],[7,4,8,3]])
df=pd.DataFrame(df,columns=['A','B','C','D'])
df
```

```python
sns.heatmap(df,annot= True)    # annot- to show numbers inside
plt.xlabel('Columns')
plt.ylabel('Rows')
```
