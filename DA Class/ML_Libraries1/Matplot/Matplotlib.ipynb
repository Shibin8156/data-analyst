### Pyplot is a matplotlib module which provides functions that interact with the figure. i.e creates a figure ,decorates the plot with labels , creates plotting area in a figure

```python
# from matplot import pyplot as lit
import matplotlib.pyplot as plt
```

```python
x=['Shyam','Aswin','Praveen','Anjana','Jubil','Shibin']
y=[86,75,76,74,72,85]
plt.plot(x,y)
```

```python
x=['Shyam','Aswin','Praveen','Anjana','Jubil','Shibin']
y=[86,75,76,74,72,85]
plt.plot(x,y)
plt.title('Person weight diagram')
plt.xlabel('Name')
plt.ylabel('weight')
```

```python
plt.figure(figsize=(8,7)) # changing size of the figure
x=['Shyam','Aswin','Praveen','Anjana','Jubil','Shibin']
y=[86,75,76,74,72,85]
plt.plot(x,y)
plt.xlim(-1,6)
plt.ylim(70,90)
plt.title('Person weight diagram')
plt.xlabel('Name')
plt.ylabel('weight')
```

```python
x1=['Shyam','Aswin','Praveen','Anjana','Jubil','Shibin']
y1=[86,75,76,74,72,85]
plt.plot(x,y)

y2=[75,75,80,65,70,78]
plt.plot(x1,y2)

plt.title('Person weight diagram')
plt.xlabel('Name')
plt.ylabel('weight')
plt.legend(['current weight','desired weight'])
```

```python
# Subplot
a=[10,20,30,40,50]
b=[1,2,3,4,5]
c=[2,3,6,8,10]
d=[3,2,1,0,6]

plt.subplot(2,2,1)  # plt.subplot(221)
plt.plot(a,b)
plt.subplot(2,2,2)  # plt.subplot(222)
plt.plot(b,c)
plt.subplot(2,2,3)  # plt.subplot(223)
plt.plot(a,c)
plt.subplot(224)  # plt.subplot(224)
plt.plot(d,b)
plt.xlabel('x_values')
plt.ylabel('y_values')
```

```python
x1=['Shyam','Aswin','Praveen','Anjana','Jubil','Shibin']
y1=[86,75,76,74,72,85]
plt.plot(x,y, label="current weight", c='green', linestyle='-.' , marker='s')
# linestyle='-','--','-.',':'
y2=[75,75,80,65,70,78]
plt.plot(x1,y2, label='desired weight' , c='aqua', linestyle=':', marker='*')

plt.title('Person weight diagram')
plt.xlabel('Name')
plt.ylabel('weight')
plt.legend()
```

```python
# Bar Graph
x4=['Shyam','Aswin','Praveen','Anjana','Jubil','Shibin']
y5=[86,75,76,74,72,85]
plt.bar(x4,y5, color='darkred')
plt.title('Person weight diagram')
plt.xlabel('Name')
plt.ylabel('weight')
```

```python
# multiple plotting
import numpy as np
plt.figure(figsize=(8,7))
xx=['Shyam','Aswin','Praveen','Anjana','Jubil','Shibin']
x_range=np.arange(len(xx)) # (0,1,2,3,4,5)
width=0.4
yy=[86,75,76,74,72,85]

#  x_range - width= (-0.4,0.6,1.6,2.6,3.6,4.6)
# end points= (0,1,2,3,4,5)
plt.bar(x_range-width, yy,width=width,label='current weight',color='brown')
yy2=[75,75,80,65,70,78]
plt.bar(x_range, yy2,width=width, label='desired weight', color='green')

plt.xticks(ticks=x_range, labels=xx)
plt.grid(True)
plt.title('Person weight diagram')
plt.xlabel('Name')
plt.ylabel('weight')
plt.legend()
```

```python
# Horizontal bar graph
# x values changes into y axis values vice versa
x=['Shyam','Aswin','Praveen','Anjana','Jubil','Shibin']
y=[86,75,76,74,72,85]
plt.barh(x,y)

plt.title('Person weight diagram')
plt.xlabel('Name')
plt.ylabel('weight')
```

```python
# pie charts
slices= [30,40,20]
labels=['men','women','kids']
plt.style.use('classic')
plt.pie(slices,labels=labels)
plt.title('PIE CHART')
```

```python
x= [30,40,20]
y=['men','women','kids']
col=['cyan','red','orange']
plt.style.use('classic')
plt.pie(x, labels=y, colors=col, wedgeprops={'linewidth': 2, 'edgecolor':'green'})
plt.title('PIE CHART')
```

```python
slices= [30,40,20]
labels=['men','women','kids']
colors=['cyan','red','orange']
explodes=[0,0,0.1]
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'linewidth': 2, 'edgecolor':'black'},
       explode= explodes, shadow=True)
plt.title('PIE CHART')
```

```python
slices= [30,40,20]
labels=['men','women','kids']
colors=['cyan','red','orange']
explodes=[0,0,0.1]
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'linewidth': 2, 'edgecolor':'purple'},
       explode= explodes, shadow=True,
       startangle=90,
       autopct='%.2f%%')

plt.title('PIE CHART')
```

```python
# Scatter plot
import numpy as np
a=np.random.randint(1,10,size=9)
b=np.random.randint(2,15,9)
plt.style.use("seaborn")
print(a)
print(b)
plt.scatter(a,b)
```

```python
import numpy as np
a=np.random.randint(1,10,size=9)
b=np.random.randint(2,15,9)
plt.style.use("fivethirtyeight")
print(a)
print(b)
plt.scatter(a,b,  s=140,color="aqua", marker='*',
            edgecolors="black", linewidth=2)
```

```python
a=np.random.randint(1,10,size=9)
b=np.random.randint(2,15,9)
colors=b
plt.style.use("tableau-colorblind10")
print(a)
print(b)
plt.scatter(a,b,  s=140, marker='s',c=colors,cmap='Purples',# blues,greens,summmer,autumn
            edgecolors="black", linewidth=2)
cbar=plt.colorbar()
cbar.set_label('color variation')
```

```python
a=np.random.randint(1,10,size=9)
b=np.random.randint(2,15,9)
print(b)
plt.style.use("tableau-colorblind10")
colors=b
sizes=b*10
plt.scatter(a,b,s=sizes, marker='s',c=colors,cmap='Purples',# blues,greens,summmer,autumn
            edgecolors="black", linewidth=2)
cbar=plt.colorbar()
cbar.set_label('color variation')
```

```python
# stack plot
days=[1,2,3,4,5]
studying=[6,8,7,6,11]
playing=[5,8,7,6,13]
plt.stackplot(days,studying,playing, colors=['orange','green'])
plt.xlabel('Days')
plt.ylabel('Number of hours')
plt.legend(['studying','playing'])
```
