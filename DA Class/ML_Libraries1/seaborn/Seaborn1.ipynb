```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('iris.csv')
data.head()
```

### line plot

```python
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
```

### styling plot

#### set_style()
  ##### 5 themes are available: darkgrid, whitegrid,dark,wite,ticks

```python
sns.set_style('white')
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
```

```python
sns.set_style('ticks')
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
sns.despine()
```

```python
sns.set_style('darkgrid')
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
```

```python
sns.set_style('whitegrid')
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
```

```python
sns.set_style('dark')
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
```

### despine()
#### Remove, borders, Removes

```python
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
sns.despine()
```

```python
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
sns.despine(left=True)
```

```python
sns.set_style('white')
plt.figure(figsize=(3,3))
sns.lineplot(x='SepalLengthCm', y= 'SepalWidthCm', data=data)
plt.title('Sepel length v/s Sepel Width')
```
